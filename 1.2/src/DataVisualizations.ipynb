```python
import seaborn as sns
```

```python
# visualization 1: Box plot of writing scores by Lunch
sns.boxplot(x='lunch', y='writing score', data=data1)
plt.title('Writing Scores by Lunch')
plt.savefig ('/Users/balajipavan/Downloads/PDS_Assignment-1/1.2/results/Writing_Scores_by_Lunch.png')
plt.show()
```

```python
# visualization 2: Strip plot of Total and Lunch
l_g= sns.stripplot(x ='lunch', y= 'Total',hue='lunch', data = data1)
l_f= l_g.get_figure()
l_f.savefig("/Users/balajipavan/Downloads/PDS_Assignment-1/1.2/results/lunchimage.png")
```

```python
# visualization 3: Bar plot of test preparation distribution
test_prep_counts = data1['test preparation course'].value_counts()
plt.bar(test_prep_counts.index, test_prep_counts.values)
plt.title('Test Preparation Distribution')
plt.xlabel('Test_Prep')
plt.ylabel('Count')
plt.savefig ('/Users/balajipavan/Downloads/PDS_Assignment-1/1.2/results/Test_Prep_distribution.png')
plt.show()
```

```python
# visualization 4: Scatter plot of writing scores vs reading scores
plt.scatter(data1['writing score'], data1['reading score'])
plt.title('Writing score vs Reading score')
plt.xlabel('writing score')
plt.ylabel('reading score')
plt.savefig ('/Users/balajipavan/Downloads/PDS_Assignment-1/1.2/results/Writing_score_vs_Reading_score.png')
plt.show()
```

```python
# visualization 5: Density plot of reading scores by gender 
sns.kdeplot(x='reading score', data=data1, hue='gender', fill=True, alpha=.3)
plt.title('Reading Scores by Gender')
plt.xlabel('Reading Score')
plt.savefig ('/Users/balajipavan/Downloads/PDS_Assignment-1/1.2/results/Reading_Scores_by_Gender.png')
plt.show()
```
